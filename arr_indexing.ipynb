## Numpy indexing and slicing

```python
import numpy as np
```

```python
arr = np.arange(1,11)
arr
```

```python
arr[6]
```

```python
arr[1:6]
```

```python
arr[:5]
```

```python
arr[5:]
```

```python
arr[2::2]
```

## Numpy indexing and slicing of matrix

```python
arr = np.arange(1,31).reshape(6,5)
```

```python
arr
```

```python
arr[0]
```

```python
arr[0][0]
```

```python
arr[4][3]
```

```python
arr[-1][1]
```

```python
arr[0,4]
```

```python
arr[5,4]
```

```python
arr[0:2,1:3]
```

```python
arr[3:6,4:6]
```

```python
arr[3:6,3:5]
```

## Boolean Indexing

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr % 2 == 0
```

```python
bool_index
```

```python
arr = arr[bool_index]
arr
```
