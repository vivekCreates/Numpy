### Valid Sudoku

```python
s = np.array([
    [5,3,4,6,7,8,9,1,2],
    [6,7,2,1,9,5,3,4,8],
    [1,9,8,3,4,2,5,6,7],
    
    [8,5,9,7,6,1,4,2,3],
    [4,2,6,8,5,3,7,9,1],
    [7,1,3,9,2,4,8,5,6],
    
    [9,6,1,5,3,7,2,8,4],
    [2,8,7,4,1,9,6,3,5],
    [3,4,5,2,8,6,1,7,9]
])
```

```python
sm  = np.sum(s,axis=1)
```

```python
sm == 45
```

```python
for num in sm:
    if num != 45:
        print("Soduku is not valid")
        break
else:
    print("Soduku is valid for rows") 
```

```python
a = np.sum(s,axis=0)
a
```

```python
for num in a:
    if num != 45:
        print("Soduku is not valid")
        break
else:
    print("Soduku is valid for colums")
```

```python
for i in range(0,9,3):
    for j in range(0,9,3):
       mat =  s[i:i+3,j:j+3]
       if np.sum(mat) != 45:
           print("Sudoku is not valid")
           break
else:
    print("Sudoku is Valid ")       
```
